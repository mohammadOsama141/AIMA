```python
import torch
from diffusers import AutoPipelineForInpainting
from diffusers.utils import load_image, make_image_grid

pipeline = AutoPipelineForInpainting.from_pretrained(
    "kandinsky-community/kandinsky-2-2-decoder-inpaint", torch_dtype=torch.float16
)
pipeline.enable_model_cpu_offload()
# remove following line if xFormers is not installed or you have PyTorch 2.0 or higher installed
pipeline.enable_xformers_memory_efficient_attention()
```

```python
img = "/content/img.png"
image = load_image(img)

mask = "/content/mask.png"
mask_image = load_image(mask)
```

```python
from PIL import Image
import matplotlib.pyplot as plt
```

```python
# Display the image using Matplotlib
plt.imshow(image)
plt.axis('off')  # Hide axes
plt.show()
```

```python
# Display the image using Matplotlib
plt.imshow(mask_image)
plt.axis('off')  # Hide axes
plt.show()
```

```python
prompt = "holding a black sharp blade, highly detailed, 8k"
negative_prompt = "bad anatomy, deformed, ugly, disfigured"
image = pipeline(prompt=prompt, negative_prompt=negative_prompt, image=image, mask_image=mask_image).images[0]
make_image_grid([image, mask_image, image], rows=1, cols=3)
```

```python
prompt = "wearing a black jacket, highly detailed, 8k"
negative_prompt = "bad anatomy, deformed, ugly, disfigured, blurry "
image = pipeline(prompt=prompt,
                 negative_prompt=negative_prompt,
                 image=image, mask_image=mask_image,
                 num_inference_steps=70).images[0]

make_image_grid([image, mask_image, image], rows=1, cols=3)
```

```python
prompt = "wearing a blue backpack, highly detailed"
negative_prompt = "bad anatomy, deformed, ugly, disfigured, blurry, blue shirt "
image = pipeline(prompt=prompt,
                 negative_prompt=negative_prompt,
                 image=image, mask_image=mask_image,
                 num_inference_steps=150).images[0]

make_image_grid([image, mask_image, image], rows=1, cols=3)
```

# inpaint using SDXL

```python
import torch
from diffusers import StableDiffusionXLInpaintPipeline
from diffusers.utils import load_image, make_image_grid
```

```python
pipe = StableDiffusionXLInpaintPipeline.from_pretrained(
    "stabilityai/stable-diffusion-xl-base-1.0",
    torch_dtype=torch.float16,
    variant="fp16",
    use_safetensors=True,
)
pipe.to("cuda")
```

```python
img = "/content/img.png"
init_image = load_image(img)

mask = "/content/mask.png"
mask_image = load_image(mask)
```

```python
prompt = "wearing a blue backpack"
image = pipeline(
    prompt=prompt, image=init_image, mask_image=mask_image, num_inference_steps=50, strength=0.80
).images[0]
```

```python
make_image_grid([image, mask_image, image], rows=1, cols=3)
```
